### Question 6

Using the `starwars` dataset, identify the individuals with:

- the tallest height
- the lowest mass

HINT: the starwars dataset is part of the dplyr package

```r
library(dplyr) #load dplyr

data("starwars") #load dataset

height <- starwars %>%
    filter(!is.na(height)) %>% #filters out NA values
    arrange(desc(height)) #sorts by descending height

tallest_character <- height[1, ]

cat("The tallest character is ", tallest_character$name, " at ", tallest_character$height, "cm\n")

lowmass <- starwars %>%
    filter(!is.na(mass)) %>%
    arrange(mass)

lightest_character <- lowmass[1, ]

cat("The lightest character is ", lightest_character$name, " at ", lightest_character$mass, "kg")
```

### Question 7

The `dplyr` dataset `storms` is part of the US National Oceanic and Atmospheric Administration database of Atlantic hurricanes. It contains multiple repeated measurements from hurricanes. For a given storm, there are measurements from every 6 hours across the life time of the storm.

Use a tidyverse approach with `dplyr` to answer the following:

- how many different storms are contained in the database?
- which year had the most storms?
- which storm, in which year had the highest pressure?

HINT:
unique() will help ensure you remove duplicated data from a vector e.g.
x <- c(1, 1, 2, 2, 3, 3, 4, 4)
unique(x)

```r
#load packages
library(dplyr)
library(tidyverse)

#load data
data("storms")

#how many different storms are contained within the database?
cat("There are ", n_distinct(storms$name), " different storms within the database \n")

#which year had the most storms?
count_year <- storms %>% count(year, sort = TRUE, name = "No_storms_per_year") #count number of storms in each year and sort them descending

most_storms <- count_year[1, ] #filter just the top result

cat("The most storms in a year was ", most_storms$No_storms_per_year, " storms, which happened in ", most_storms$year, "\n")

#which storm, in which year, had the highest pressure

press_sort <- arrange(storms, desc(pressure)) #arrange the dataset in descending order of pressure

highest_press <- press_sort[1, ] #select the highest pressure value

cat("The highest pressure was ", highest_press$pressure, "Pa in the year ", highest_press$year)
```

### Question 8

The `ggplot2` dataset `diamonds` contains the price and information from the sale of 54,000 different diamonds. You can use ?diamonds to find out more about the dataset if you want.

Using `ggplot2`:

- plot the relationship between the carat (i.e. the weight of the diamond) and the price.
- plot a histogram of prices of all the diamonds
- plot a boxplot of the difference in price between the different cuts

```r
library(tidyverse) #load tidyverse which includes ggplot2

data("diamonds") #load the diamonds dataset

#plot the relationship between the carat and the price
ggplot(data = diamonds, aes(x = carat, y = price)) +
    geom_point() + #declare graph type to be scatter
    theme_classic() + #set theme
    labs(x = "Carat", y = "Price (US$)", title = "Carat Value vs Price of Diamonds") + #rename axis and add title
    theme(plot.title = element_text(hjust = 0.5, size = 24)) #set title to centre and increase size

#plot a histograms of prices of all the diamonds
ggplot(data = diamonds, aes(x = price)) +
    geom_histogram(binwidth = 2000, aes(fill = "pink")) + #set the graph type to histogram, the size of each bin to 2000, and the colour to pink
    theme_minimal() + #set the theme
    labs(x = "Price (US$)", y = "Frequency", title = "Histogram of diamond prices") + #rename the axis and add a title
    theme(plot.title = element_text(hjust = 0.5, size  = 24))+ #set title to centre and increase size
    guides(fill = FALSE) #remove legend

#plot a boxplot of the difference in price between the different cuts
ggplot(data=diamonds, aes(x = cut, y = price, fill = cut)) + #set dataset and values
    geom_boxplot() + #set graph type
    theme_minimal() + #set theme
    labs(x = "Cut", y = "Price (US$)", title = "Box plot of prices for different cuts of diamond") + #label axis and add title
    theme(plot.title = element_text(hjust = 0.5, size = 24)) + #set title location and size
    scale_y_continuous(limits = quantile(diamonds$price, c(0.1, 0.9))) #trim the top 10% to make graph more readable
```
